```python
import itertools
import numpy as np
import shutil
from pathlib import Path
```

```python
f = lambda x: x%2
a = itertools.filterfalse(f,range(10))
for element in a:
    print(element)
```

```python
l = ['A','E', 'I', 'O', 'U']

f = lambda x: x[0] in l

names = ['Ferdinand', 'Alan', 'Mônica', 'Jéssica', 'Eduardo','Felipe', 'Pedro','André']

vogais = itertools.filterfalse(f, names)
for name in vogais:
    print(name)
```

```python
n = itertools.count(start=0, step=1)
next(n)
next(n)
yield (n)
```

```python
a = [('Felipe','Veiga'),('Felipe','Marinho'),('Eduardo','Veiga')]
d = itertools.groupby(a, key = lambda x: x[0])
for label, groups in d:
    print({label: len(list(groups))}) 
```

```python
for i in itertools.islice(range(10),1, 8): 
    print(i)
```

```python
np.cumsum([1,2,3,4,5])
```

```python
a = itertools.cycle([1,2,3])
for element in a:
    print(element)
```
