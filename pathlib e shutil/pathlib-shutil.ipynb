```python
# Uma maneira mais sofisticada de se mover os arquivos que baixo para a pasta de política
from pathlib import Path
import shutil
import os
import datetime 
hoje = datetime.datetime.now().date()
origin = Path.cwd().parents[3] /'Downloads'
destiny = Path.cwd().parents[2]/'Política'

for file in origin.iterdir():
    today = datetime.datetime.fromtimestamp(os.stat(file).st_birthtime).date()
    year = datetime.datetime.fromtimestamp(os.stat(file).st_birthtime).strftime('%Y')                                                                  
    month = datetime.datetime.fromtimestamp(os.stat(file).st_birthtime).strftime('%B')
    if today == hoje:
        for direct_year in destiny.iterdir():
            if direct_year.name == year:
                for direct_month in direct_year.iterdir():
                    if direct_month.name == month:
                        shutil.move(str(file), str(direct_month)) 
    
    
```
